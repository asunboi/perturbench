```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define axes labels
x_labels = ["low", "medium", "high"]      # training splits
y_labels = ["high", "medium", "low"]      # data quality

# Create empty DataFrame with zeros
data = pd.DataFrame(0, index=y_labels, columns=x_labels)

# Plot
plt.figure(figsize=(6,5))
ax = sns.heatmap(
    data, annot=False, cbar=False, 
    linewidths=1, linecolor='black',
    square=True, cmap="Greys"
)

ax.set_xlabel("Training Data Amount")
ax.set_ylabel("Training Data Quality")
plt.title("Predicted vs Observed Pearson Correlation")

plt.show()
```
